```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv("/content/sample_data/car dekkho.csv")

# Extract the required data for x and y axes
x = [2010, 2011, 2012, 2013, 2014]  # x-axis values (time)
y = data.groupby('year')['km_driven'].sum().loc[x]  # y-axis values (dependent variable)

# Create a line chart
plt.plot(x, y)

# Customize the chart
plt.title("Line Chart Example")
plt.xlabel("Year")
plt.ylabel("Total Kilometers Driven")

# Display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/car dekkho.csv')

# Get the top 5 values based on mean selling price
top_5 = df.groupby('year')['selling_price'].mean().nlargest(5)

# Extract x-axis values (name) and y-axis values (mean selling price) for the top 5
x = top_5.index
y = top_5.values

# Create a line chart
plt.plot(x, y)

# Customize the chart
plt.title("Top 5 Automobile Analysis")
plt.xlabel('year')
plt.ylabel('Mean Selling Price')

# Display the chart
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/car dekkho.csv')

# Grouping the data by year and summing the values
data = df.groupby('year')['selling_price'].sum()

# Plotting the histogram
plt.hist(data, bins=5, edgecolor='black')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Displaying the histogram
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/car dekkho.csv')

# Assuming 'df' is your DataFrame containing the relevant data

top_10 = df.groupby('fuel').head(10)

grouped_data = top_10.groupby('fuel')

plt.figure()  # Create a new figure

for name, group in grouped_data:
    plt.plot(group['year'], group['km_driven'], label=name)

plt.xlabel('Year')
plt.ylabel('Kilometres Driven')
plt.title('Line Chart')
plt.legend()
plt.show()




```

```python
grouped_data = df.groupby('transmission')
for name, group in grouped_data:
    plt.hist(group['fuel'], label=name, alpha=1)

plt.xlabel('Fuel Type')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.legend()
plt.show()
```
